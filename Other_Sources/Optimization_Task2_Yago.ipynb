# Custom Class for Linear SVM

```python
import numpy as np

class SimpleSVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize weights and bias
        self.w = np.zeros(n_features)
        self.b = 0

        y_ = np.where(y <= 0, -1, 1)  # Convert labels to -1 and 1

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                condition = y_[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    self.w -= self.learning_rate * (2 * self.lambda_param * self.w)
                else:
                    self.w -= self.learning_rate * (2 * self.lambda_param * self.w - np.dot(x_i, y_[idx]))
                    self.b -= self.learning_rate * y_[idx]

    def predict(self, X):
        approx = np.dot(X, self.w) - self.b
        return np.sign(approx)



svm = SimpleSVM()
svm.fit(X_train_scaled, y_train)
y_pred = svm.predict(X_test_scaled)
y_pred = np.where(y_pred == -1, 0, y_pred)
y_pred = y_pred.astype(int)
```

# Custom Class for Non-Linear SVM With Kernel Trick


```python
def polynomial_kernel(x1, x2, degree=3, coef0=1):
    return (np.dot(x1, x2) + coef0) ** degree

def rbf_kernel(x1, x2, gamma=0.1):
    return np.exp(-gamma * np.linalg.norm(x1 - x2) ** 2)

class KernelSVM:
    def __init__(self, kernel=rbf_kernel, learning_rate=0.001, lambda_param=0.01, n_iters=1000):
        self.kernel = kernel
        self.learning_rate = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.alpha = None
        self.b = 0
        self.X_train = None
        self.y_train = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize alpha (Lagrange multipliers)
        self.alpha = np.zeros(n_samples)
        self.b = 0

        # Store the training data
        self.X_train = X
        self.y_train = np.where(y <= 0, -1, 1)  # Convert labels to -1 and 1

        # Precompute the kernel matrix
        K = np.zeros((n_samples, n_samples))
        for i in range(n_samples):
            for j in range(n_samples):
                K[i, j] = self.kernel(X[i], X[j])

        for _ in range(self.n_iters):
            for idx in range(n_samples):
                condition = self.y_train[idx] * (np.sum(self.alpha * self.y_train * K[:, idx]) - self.b) >= 1
                if condition:
                    self.alpha[idx] -= self.learning_rate * (2 * self.lambda_param * self.alpha[idx])
                else:
                    self.alpha[idx] -= self.learning_rate * (2 * self.lambda_param * self.alpha[idx] - self.y_train[idx])
                    self.b -= self.learning_rate * self.y_train[idx]

    def project(self, X):
        y_predict = np.zeros(X.shape[0])
        for i in range(X.shape[0]):
            s = 0
            for alpha, x_train, y_train in zip(self.alpha, self.X_train, self.y_train):
                s += alpha * y_train * self.kernel(X[i], x_train)
            y_predict[i] = s
        return y_predict

    def predict(self, X):
        return np.sign(self.project(X) - self.b)
```
